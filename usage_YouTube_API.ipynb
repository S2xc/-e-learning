# YouTube API
У YouTube API есть ограниченное на количество квот, подробнее: https://developers.google.com/youtube/v3/getting-started?hl=ru#quota

# Что нужно знать перед созданием проекта
Выдержа из инструкции <u>**Использование квоты**</u>: 

*Проекты, в которых используется API данных YouTube, имеют квоту по умолчанию в размере 10 000 единиц в день — сумму, достаточную для подавляющего большинства пользователей нашего API.*

Выдержа из инструкции <u>**Расчет использования квоты**</u>: 

*Google рассчитывает использование квоты, назначая стоимость каждому запросу. Разные виды операций имеют разную стоимость квот. Например:*

+ *Операция чтения, извлекающая список ресурсов — каналов, видео, плейлистов — обычно стоит 1 единицу.*
+ *Операция записи, которая создает, обновляет или удаляет ресурс, обычно стоит 50 единиц.*
+ *Поисковый запрос стоит 100 единиц.*
+ *Загрузка видео стоит 1600 единиц.*

# Чтобы получить API ключ для использования YouTube Data API, нужно выполнить следующие шаги:

## Шаг 1: Создание проекта на Google Developers Console
+ Перейдите на Google Developers Console.
+ Войдите в свою учетную запись Google или создайте новую.
+ Нажмите на кнопку "Создать проект" или выберите существующий проект из выпадающего списка.

## Шаг 2: Включение YouTube Data API
+ После создания проекта, выберите его из списка.
+ В меню слева выберите "Библиотека".
+ В разделе "YouTube API" найдите и выберите "YouTube Data API v3".
+ Нажмите "Включить" (Enable).

## Шаг 3: Создание учетных данных (API ключ)
+ В меню слева выберите "Учетные данные" (Credentials).
+ Нажмите на кнопку "Создать учетные данные" (Create credentials) и выберите "API ключ" (API key).
+ Ваш API ключ будет создан и показан. Скопируйте его и сохраните в безопасном месте.
